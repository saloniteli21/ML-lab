```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
unemployment = pd.read_csv("/content/sample_data/clean_unemployment.csv")
```

```python
unemployment.head()
```

```python
unemployment.describe()
```

```python
unemployment.info()
```

```python
unemployment.mean().plot(xlabel='Year', ylabel='Unemployment Rate', xticks=np.arange(0,12), grid=True, figsize=(10,5))
```

```python
unemployment[unemployment['continent']=='Asia'].mean().plot()
unemployment[unemployment['continent']=='Europe'].mean().plot()

unemployment[unemployment['continent']=='South America'].mean().plot(figsize=(12,6),xticks=np.arange(0,12))
plt.legend(['Asia','Europe','South America'], loc='upper right', ncols=2, fontsize=8)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Yearly Mean Unemployment Rate by Continent')
```
